Feature Encoding


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df=pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['CustomerId','RowNumber','Surname'],axis=1,inplace=True)
```

```python
df.head()
```

Label Encoding

```python
le=preprocessing.LabelEncoder()
df['Gender_label']=le.fit_transform(df.Gender.values)
```

```python
df.head(10)
```

One Hot Encoding

```python
one_hot=pd.get_dummies(df['Geography']).astype(int)
```

```python
one_hot
```

```python
df_dummies=pd.get_dummies(df)
```

```python
df_dummies.head()
```

```python
df_dummies_de=pd.get_dummies(df,drop_first=True)
df_dummies_de.head(5)
```
