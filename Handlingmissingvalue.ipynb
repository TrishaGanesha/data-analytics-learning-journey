Handling Missing Values

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("synthetic_dataset[1].csv")
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
#handling missing values
#1.deleting the columns with missing data but we have almost all col with missing data
#for col updated_df=df.dropna(axis=1)
#2.deleting the rows with missing data
```

```python
updated_df=df.dropna(axis=0)
```

```python
updated_df.info()
```

```python
#3.Imputation
#filling with mean(less outliners) and median(more outliners) values if numerical ,mode if categorical data
```

```python
df['Price'].mean()
```

```python
df['Rating'].mean()
```

```python
df['Discount'].mean()
```

```python
df['Category'].mode()
```

```python
df['Stock'].mode()
```

```python
updated_df=df
updated_df['Price']=updated_df['Price'].fillna(df['Price'].mean())
updated_df.info()
```

```python
updated_df=df
updated_df['Rating']=updated_df['Rating'].fillna(df['Rating'].mean())
updated_df.info()
```

```python
updated_df=df
updated_df['Discount']=updated_df['Discount'].fillna(df['Discount'].mean())
updated_df.info()
```

```python
updated_df=df
updated_df['Category']=updated_df['Category'].fillna(df['Category'].mode()[0])
updated_df.info()
```

```python
updated_df=df
updated_df['Stock']=updated_df['Stock'].fillna(df['Stock'].mode()[0])
updated_df.info()
```
