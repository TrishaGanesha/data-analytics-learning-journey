```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)

df.head()
```

```python
df.info()
```

#for standardization and Normalization
#standardScaler(features removing the mean and scaling to unit variance,zscore z=(x-mean)/std deviation) 
#MinMaxScaler(transform features by scaling each to a given range)
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

#Normalization=scaling technique in which values are shifted and rescaled between 0 and 1 also called as Min Max scaling
#Standardization=scaling technique where values are centered around the mean with a unit std deviation.This means that the mean of the attribute becomes zero and the resultant distribution has a unit std deviation


```python
df.describe().round(2)#statistical information
```

Task1 Normalization

```python
df.head(5)
```

```python
scaler=MinMaxScaler() #Instantiating the MinMax() function
normalized_df=scaler.fit_transform(df)
print(normalized_df)
```

Task 2 Standardization

```python
scaler=StandardScaler()
normalized_df=scaler.fit_transform(df)
print(normalized_df)
```

```python
import numpy as np

# Check mean and variance of the scaled data
print("Means:", np.mean(normalized_df, axis=0))
print("Variances:", np.var(normalized_df, axis=0))
```
